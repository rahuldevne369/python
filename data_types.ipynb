Python has several built-in data types, each designed for specific kinds of data and operations. Here are the most common ones with examples:

Text Type (str)
Represents sequences of characters enclosed in quotes.

Example: text = "Hello, Python!" 
Numeric Types (int, float, complex)

int: Whole numbers (positive or negative).
Example: integer_value = 42

float: Decimal numbers.
Example: float_value = 3.14

complex: Numbers with a real and imaginary part (using j for the imaginary unit).
Example: complex_value = 2 + 3j 

Sequence Types (list, tuple, range)

list: Ordered, mutable collections.
Example: my_list = [1, 2, 3]

tuple: Ordered, immutable collections.
Example: my_tuple = (4, 5, 6)

range: Generates a sequence of numbers.
Example: my_range = range(1, 5) 
Mapping Type (dict)
Stores data in key-value pairs.

Example: my_dict = {"name": "Alice", "age": 25} 
Set Types (set, frozenset)

set: Unordered collection of unique elements.
Example: my_set = {1, 2, 3}

frozenset: Immutable version of a set.
Example: my_frozenset = frozenset([4, 5, 6]) 
Boolean Type (bool)
Represents truth values: True or False.

Example: is_python_fun = True 
Binary Types (bytes, bytearray, memoryview)

bytes: Immutable sequence of bytes.
Example: byte_data = b"hello"

bytearray: Mutable sequence of bytes.
Example: byte_array_data = bytearray([65, 66, 67])

memoryview: Allows efficient manipulation of binary data.
Example: memory_view_data = memoryview(byte_array_data) 

None Type (NoneType)
Represents the absence of a value.

Example: no_value = None
